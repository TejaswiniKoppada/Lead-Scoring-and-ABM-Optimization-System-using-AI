```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('data.csv')
df.head()
```

```python
df.corr()
```

```python
sns.countplot(x='Cured', data=df)
plt.title('Frequency of Target Variable')
plt.show()
```

```python
df.isnull()
```

```python
y = df['Cured']
X = df.drop('Cured',axis=1)
```

```python
random_state=0
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.svm import LinearSVC
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train_std=sc_X.fit_transform(X_train)
X_test_std=sc_X.transform(X_test)
svm = LinearSVC(C=1.0, random_state=0)  # Adjust C as needed
svm.fit(X_train_std, y_train)
```

```python
from sklearn.metrics import f1_score
y_train_pred=svm.predict(X_train_std)
y_test_pred=svm.predict(X_test_std)
f1_score_train= f1_score(y_train, y_train_pred)
f1_score_test=  f1_score(y_test,y_test_pred)
print("F1 Score Train:",f1_score_train)
print("F1 Score Test:",f1_score_test)
```
