This dataset includes 200 rows with three columns: Temperature, Sales, and Advertisement. Here's an explanation of the features in the extended dataset:

Temperature: This feature represents the temperature on a particular day. It is a continuous numerical value that measures the temperature in degrees.

Sales: This feature represents the sales of a product on a specific day. It is also a continuous numerical value that measures the number of units sold.

Advertisement: This feature indicates whether an advertisement was run on that day (1) or not (0). It is a categorical binary variable that captures whether marketing efforts were made on that day.

##### 2) Let's import the dataset `icecreamsales.csv` and explore it. Store the dataset in `df`. Check the structure of the data by inspecting the number of rows, columns, and attributes of the dataset. You can use functions like .shape, .head(), .info(), and .describe() to get a quick overview of the data.

```python
df=pd.read_csv('icecreamsales.csv')
df.head()
```

```python
X = df[[ 'Temperature']]
y = df['Revenue']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred= lr.predict(X_test) 
```

```python
plt.scatter(y_pred, y_test, color='blue', label='1:1 relation')
plt.ylabel('oberved')
plt.xlabel('predicted')
plt.title('Regression analysis')
plt.legend()
plt.show()
```

##### Save the trained model to a file.

```python
dump(lr, 'linear_regression_model.joblib')
```

#####  Now your Linear Regression model is saved as a file named "linear_regression_model.joblib". You can load it back later using the load function:

```python
loaded_model = load('linear_regression_model.joblib')
```
