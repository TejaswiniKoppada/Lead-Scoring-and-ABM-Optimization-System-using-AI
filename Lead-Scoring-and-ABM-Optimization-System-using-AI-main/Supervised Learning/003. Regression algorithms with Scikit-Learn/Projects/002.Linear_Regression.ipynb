In this lab, we will shift our focus to Regression problems, another important task within Machine Learning. While classification aims to predict categorical class labels, regression involves predicting continuous numerical values based on past observations. Unlike binary classification, regression allows for an infinite number of possible output values.

We will delve into regression analysis and predict continuous variables. We will practice how to build regression models using supervised learning algorithms.

Simulated dataset
We will start by introducing a synthetic dataset. Let's use sci-kit-learn's **make_regression** function to generate a simulated dataset. For this step, we should specify the number of samples, features, noise level, and other parameters.

Let's run the code in the notebook.

```python
import numpy as np
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
X, y = make_regression(n_samples=100, n_features=1, noise=10)
```

```python
plt.scatter(X, y)
plt.xlabel('X')
plt.ylabel('y')
plt.title('Simulated Dataset')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
random_state=0
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=random_state)
```

## Linear regression
Let's utilize a linear regression algorithm.

The equation of linear regression can be summarized as  **y=mx+b**
, where y represents the dependent variable, x represents the independent variable, m represents the slope of the line, and b represents the y-intercept.

The linear regression model is a fundamental machine learning algorithm used for predicting continuous numeric values based on the relationship between independent variables (features) and the dependent variable (target). It assumes a linear relationship between the features and the target variable, aiming to find the best-fit line that minimizes the distance between the predicted values and the actual values.

```python
lr= LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
plt.scatter(X_test, y_test, color='blue', label='Actual')
plt.plot(X_test, y_pred, color='red', linewidth=2, label='Predicted')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Regression Predictions')
plt.legend()
plt.show()
```
