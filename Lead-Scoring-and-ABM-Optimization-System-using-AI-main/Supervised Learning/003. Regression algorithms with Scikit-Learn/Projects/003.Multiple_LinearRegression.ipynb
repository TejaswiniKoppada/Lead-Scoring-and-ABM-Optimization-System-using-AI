### Multiple Linear Regression : CO2 emissions in vehicles
In this type of linear regression model, the relationship between a dependent variable and two or more independent variables is examined. It is used to assess the strength of the relationship between the variables and predict the value of the dependent variable based on the values of the independent variables.

The equation for multiple linear regression can be expressed as:

**y=B0+B1X1+B2X2+.......+BnXn**

where:

y is the dependent variable or the target variable that we want to predict.
β₀ is the y-intercept or the constant term.
β₁, β₂, ..., βₚ are the coefficients or weights associated with each independent variable.
x₁, x₂, ..., xₚ are the independent variables or features used to predict the target variable.
In multiple regression, we consider p independent variables (features) to predict the target variable y. The equation represents a linear combination of the independent variables, weighted by their respective coefficients, and added to the constant term β₀. By adjusting the values of the coefficients, the model can find the best-fit line that minimizes the difference between the predicted values and the actual values of the target variable.

To use multiple regression with this library, you will first need to create a LinearRegression object, then train the model on your data using the fit() method. Once the model is trained, you can use the predict() method to make predictions on new data.

#### Dataset
The following table provides an overview of various features related to CO2 emissions in vehicles. Each feature represents a specific aspect that contributes to understanding and assessing the environmental impact of motor vehicles. These features play a crucial role in analyzing and comparing the carbon footprint and performance of different vehicles.

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('Car_CO2.csv')
df.head()
```

```python
X= df[['Volume','Weight']]
y= df['CO2']
```

```python
sns.pairplot(df, x_vars=['Weight', 'Volume'], y_vars='CO2', height=4, aspect=1, kind='scatter')
plt.show()
```

```python
sns.heatmap(df[['Weight', 'Volume','CO2']].corr(), annot = True, cmap = 'coolwarm')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred= lr.predict(X_test)
```

```python
plt.scatter(y_pred, y_test, color='blue', label='1:1 relation')
plt.ylabel('observed')
plt.xlabel('predicted')
plt.title('Regression analysis')
plt.legend()
plt.show()
```
